# 4.8.6 The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:
    If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
    If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
#Creating column for average product price.
df_op_merged['mean_product_price'] = df_op_merged.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
#View user ID and average product price.
df_op_merged[['user_id', 'mean_product_price']].head(20)
```

```python
#Creating flag for Low spender
df_op_merged.loc[df_op_merged['mean_product_price'] <10, 'spender_flag'] = 'Low'
```

```python
#Creating flag for High spender
df_op_merged.loc[df_op_merged['mean_product_price'] >= 10, 'spender_flag'] = 'High'
```

```python
df_op_merged['spender_flag'].value_counts(dropna = False)
```

# 4.8.7 In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:
    If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
    If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
    If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.” 

```python
df_op_merged['Customer_frequency'] = df_op_merged.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
df_op_merged[['user_id','Customer_frequency']].head(20)
```

```python
#Creating flag for Frequent customer
df_op_merged.loc[df_op_merged['Customer_frequency'] <= 10, 'frequency_flag'] = 'Frequent'
```

```python
#Creating flag for Regular Customer
df_op_merged.loc[(df_op_merged['Customer_frequency'] > 10) & (df_op_merged['Customer_frequency'] <= 20), 'frequency_flag'] = 'Regular'
```

```python
#Creating flag for Non-Frequent Customer
df_op_merged.loc[df_op_merged['Customer_frequency'] > 20, 'frequency_flag'] = 'Non-Frequent'
```

```python
df_op_merged['frequency_flag'].value_counts(dropna = False)
```

```python
df_op_merged[['user_id', 'Customer_frequency', 'frequency_flag']].head(20)
```
