# Creating subset.

```python
#Adding If-Then function.
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
#Applying the If-Then function to column.
df['price_range'] = df.apply(price_label, axis=1)
```

```python
#Finding count of Low/Mid/High range items
df['price_range'].value_counts(dropna = False)
```

```python
#Finding Max amount of Prices column.
df['prices'].max()
```

# 4.7.2 Suppose your clients have changed their minds about the labels you created in your “busiest_day” column. Now, they want “Busiest day” to become “Busiest days” (plural). This label should correspond with the two busiest days of the week as opposed to the single busiest day. At the same time, they’d also like to know the two slowest days. Create a new column for this using a suitable method.

```python
#Creating new Result list with 0/1 = Busiest Days, 4/3 = Least Busiest Days, All other numbers = Regular Days
result_2 = []

for value in df_order_prod_merge["orders_day_of_week"]:
  if value == 0 or value == 1:
    result_2.append("Busiest days")
  elif value == 4 or value == 3:
    result_2.append("Least busiest days")
  else:
    result_2.append("Regularly busy days")
```

```python
result_2
```

# 4.7.3 Check the values of this new column for accuracy. Note any observations in markdown format.

```python
df_order_prod_merge['busiest_days'] = result_2
```

```python
df_order_prod_merge['busiest_days'].value_counts(dropna = False)
```

```python
df_order_prod_merge['orders_day_of_week'].value_counts(dropna =False)
```

From the above, we can see the total amount of Busiest days is 11,864,412. If we add the count of the orders_days_of_week column for "0" and "1" we get 11,864,412. The numbers "4" and "3" should also equal to the least busiest days with every other day equaling the regularly busy days count.

# 4.7.4 When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day.”


```python
df_order_prod_merge['order_hour_of_day'].value_counts(dropna =False)
```

```python
#We will be dividing Most Orders for the top 8 hours, Average orders for the middle 8 and Fewest Orders for the last 8.
result_3 = []
for value in df_order_prod_merge["order_hour_of_day"]:
    if value in [10,11,14,15,13,12,16,9]:
        result_3.append ("Most Orders")
    elif value in [17,8,18,19,20,7,21,22]:
        result_3.append ("Average Orders")
    else:
        result_3.append ("Fewest Orders")
```

```python
result_3
```

```python
df_order_prod_merge['busiest_period_of_day'] = result_3
```
